```python
example = {'https://hdl.handle.net/20.500.11840/146553':
    {'title': 'Kartonnen doosje met poeder, vermoedelijk voor magisch bad balsem met olijf',
  'desc': 'Doosje balsem met een (magisch) poeder erin, dat waarschijnlijk gebruikt kan worden voor een ritueel bad om aantrekkelijk voor anderen te worden, gezien de tekst "po\' atrativo". De opdruk in rood en zwart is een afbeelding van de vrouwelijke Exu Pomba Gira. Exu is een soort duivelsfiguur in de Afro-Braziliaanse religie umbanda (en candomblé) en bemiddelaar tussen de mensen en goden (orixas). <BR>    <BR>    Pomba Gira, de vrouwelijke pendant van Exu, gedraagt zich tijdens haar verschijnen - door middel van het medium, dat door haar wordt bezeten tijdens de trance - als een prostituee, agressief, sardonisch lachend en grote hoeveelheden cachaça (suikerrietjenever) drinkend. Zij maakt avances naar mannen én vrouwen. Ze werkt ten goede en ten kwade; zij kan bijvoorbeeld ziektes genezen maar ook veroorzaken, huwelijken arrangeren maar ook ontwrichten. Zij houdt van parfums, lange rokken, dansen en lachen. De gelovigen offeren haar het vlees van een zwarte geit.<BR>    <BR>    Vervaardiging<BR>    Fabrieksprodukt van de Fabrica de defumadores "Wilson" ltda. in Sao Paolo.'},
    'handle_test': {'title':'Balsem', 'desc':'nnn nnn balsem nnn'}}
```

```python
# {"query_str":[{"handle":{"title":"","descr":"","found_in":[]}}]}
```

```python
query_results = {}
```

```python
for query_str in df['translation']:
    
    clean_query_str = str(query_str).replace('*','').replace(')','').replace('(','').replace('&#39;',"'")
    
    print(clean_query_str)
    
    #matches = []
    match_dict = {}
    
    # iterate over NMVW
    
    for handle, values in example.items():
        
        found_in = []
    
        # searching in title
        if re.search(f"\\b{clean_query_str}\\b",values['title'],re.IGNORECASE) != None:
            found_in.append('title')
        # searching in descr
        if re.search(f"\\b{clean_query_str}\\b",values['desc'],re.IGNORECASE) != None:
            found_in.append('desc')

        if len(found_in) > 0:
            match_dict[handle] = {"title":values['title'],"desc":values['desc'],"found_in":found_in}
    
    if len(match_dict) > 0:
        query_results[clean_query_str] = match_dict
    
        
    #break
```

```python
for q, results in query_results.items():
    print(q,results)
```
