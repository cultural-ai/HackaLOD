```python
import json
import rdflib
import os
from rdflib import Graph
from rdflib.namespace import Namespace, RDF, DC, DCTERMS
import pandas as pd
import string
```

```python
dc11 = Namespace("http://purl.org/dc/elements/1.1/")
```

```python
nmvw = Graph()
```

```python
nmvw_json = {}
```

```python
query_str = """

SELECT ?handle ?title ?desc

WHERE {

    ?handle dc11:title ?title .
    ?handle dc11:description ?desc .

}

"""
```

```python
for number in range(1,1571):
    filename = '/Users/anesterov/WM/XMLRDFobjects-2/ObjectsRDFXML_' + str(number) + '.rdf'
    nmvw.parse(filename, format="xml")
    query_results = nmvw.query(query_str,initNs={'dc11': dc11})
    
    for e in query_results.bindings:
        nmvw_json[str(e['handle'])] = {"title":str(e['title']), "desc":str(e['desc'])}
        
    print(number)
```

```python
with open('/Users/anesterov/reps/nmvw.json', 'w') as jf:
    json.dump(nmvw_json, jf)
```
